# Importing Python Libraries

```python
import glob
import numpy
import matplotlib
import pandas as pd
import matplotlib.pyplot
import numpy as np
```

# Create one function named visualize()

```python
def visualize(filename):
    
    # load the dataset
    data = np.loadtxt(fname = filename, delimiter=',')

    fig = matplotlib.pyplot.figure(figsize=(10.0, 3.0))

    # add sub plots to the figure
    plot1 = fig.add_subplot(1, 3, 1)
    plot2 = fig.add_subplot(1, 3, 2)
    plot3 = fig.add_subplot(1, 3, 3)
    
    #  produce a sub plot of average of each features
    plot1.set_ylabel('Average')
    plot1.set_xlabel('Feature#')
    plot1.plot(np.mean(data, axis=0))
    
    #  produce a sub plot of max of each features
    plot2.set_ylabel('Max')
    plot2.set_xlabel('Feature#')
    plot2.plot(np.max(data, axis=0))

    #  produce a sub plot of min of each features
    plot3.set_ylabel('Min')
    plot3.set_xlabel('Feature#')
    plot3.plot(np.min(data, axis=0))

    fig.tight_layout()
    matplotlib.pyplot.show()
```

# Create one function named identify_issues()

```python
def identify_issues(file):
    
    # load the dataset
    data = pd.read_csv(file)
    
    # Use conditional statements to identidy issues and display the related message
    if np.max(data, axis=0)[0] == 0 and np.max(data, axis=0)[20] == 20:
        print("I doubt the integrity of this dataset")
    elif  sum(np.min(data, axis=0)) == 0:
        print("Why is this happening?")    
    else:
        print("Dataset looks good")
```

# Use one for loop to visualize the datasets and identify the related issues

```python
dataset = sorted(glob.glob('data*.csv'))

for datasets in dataset[:4]:
    print("\n\n\n\n")
    print(datasets)
    visualize(datasets)
    identify_issues(datasets)
    
```
